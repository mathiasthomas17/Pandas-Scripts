SORTING - Both a DF and a Series Method

```python
happiness
```

```python
#sort_values ->Asceding By Default

happiness.sort_values('Healthy life expectancy')#Does not Mutate DF
```

```python
happiness.sort_values('Healthy life expectancy',ascending=False) #Does not Mutate DF
```
