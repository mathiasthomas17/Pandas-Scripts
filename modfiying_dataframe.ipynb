Dropping Columns

```python
# Returns a new DF
# Does not Mutate Original
BTC.drop(labels='Symbol',axis=1)
```

```python
# Dropping Multiple
BTC.drop(labels=['Symbol','SNo','Name'],axis='columns')
```

```python
# Using drop Cols instead of labels
BTC.drop(columns=['Name','Symbol'])
```

Dropping ROws

```python
BTC.set_index('Name')
```

```python
BTC.drop(labels='Bitcoin',axis=0)
```
